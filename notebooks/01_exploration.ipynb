```python
import pandas as pd
df = pd.read_csv("data/application_train.csv")
print(df.shape)
```

```python
print(df['TARGET'].value_counts())
print(df['TARGET'].value_counts(normalize = True))
```

```python
print(df.columns.tolist())
```

```python
df.head()
```

```python
key_cols = ['SK_ID_CURR', 'TARGET', 'NAME_CONTRACT_TYPE', 'AMT_INCOME_TOTAL', 
            'AMT_CREDIT', 'AMT_ANNUITY', 'NAME_EDUCATION_TYPE', 
            'NAME_FAMILY_STATUS', 'DAYS_EMPLOYED', 'DAYS_BIRTH',
            'CNT_CHILDREN', 'OCCUPATION_TYPE']

df[key_cols].head(10)
```

```python
print(df['DAYS_BIRTH'].head() / -365)
```

```python
df['DTI_RATIO'] = df['AMT_ANNUITY'] / df['AMT_INCOME_TOTAL']
print(df['DTI_RATIO'].describe())
```

```python

import matplotlib.pyplot as plt

defaulters = df[df['TARGET'] == 1]['DTI_RATIO']
non_defaulters = df[df['TARGET'] == 0]['DTI_RATIO']

print("Avg DTI - Defaulters:", round(defaulters.mean(), 4))
print("Avg DTI - Non-defaulters:", round(non_defaulters.mean(), 4))
```

```python
high_dti = df[df['DTI_RATIO'] > 0.5]
low_dti = df[df['DTI_RATIO'] <= 0.5]

print("Default rate when DTI > 50%:", round(high_dti['TARGET'].mean() * 100, 2), "%")
print("Default rate when DTI <= 50%:", round(low_dti['TARGET'].mean() * 100, 2), "%")
print("Count with DTI > 50%:", len(high_dti))
```

```python
df['AGE_YEARS'] = df['DAYS_BIRTH'] / -365

young = df[df['AGE_YEARS'] < 30]
middle = df[(df['AGE_YEARS'] >= 30) & (df['AGE_YEARS'] < 50)]
senior = df[df['AGE_YEARS'] >= 50]

print("Default rate - Under 30:", round(young['TARGET'].mean() * 100, 2), "%")
print("Default rate - 30 to 50:", round(middle['TARGET'].mean() * 100, 2), "%")
print("Default rate - Over 50:", round(senior['TARGET'].mean() * 100, 2), "%")
```

```python
print("=== EDUCATION ===")
print(df.groupby('NAME_EDUCATION_TYPE')['TARGET'].mean().sort_values(ascending=False).round(4) * 100)

print("\n=== FAMILY STATUS ===")
print(df.groupby('NAME_FAMILY_STATUS')['TARGET'].mean().sort_values(ascending=False).round(4) * 100)

print("\n=== CONTRACT TYPE ===")
print(df.groupby('NAME_CONTRACT_TYPE')['TARGET'].mean().sort_values(ascending=False).round(4) * 100)
```

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100
missing_report = missing_pct[missing_pct > 0].sort_values(ascending=False)
print(f"Columns with missing values: {len(missing_report)} out of {len(df.columns)}")
print(missing_report.head(15))
```
